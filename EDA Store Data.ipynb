## **HYPOTHESIS TESTING**

## **Hypothesis Testing 1**

```python
from scipy import stats

# Assuming you have a single DataFrame named 'sales_data'

# Separate sales data into two groups based on gender
sales_men = sd[sd['Gender'] == 'Men']['Amount']
sales_women = sd[sd['Gender'] == 'Women']['Amount']

# Perform an independent t-test to compare sales between men and women
t_statistic, p_value = stats.ttest_ind(sales_men, sales_women, equal_var=False)

# Define the significance level (alpha)
alpha = 0.05

# Check the p-value against the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
    print("There is a significant difference in sales between men and women.")
else:
    print("Fail to reject the null hypothesis")
    print("There is no significant difference in sales between men and women.")
```

## **Hypothesis Testing 2**

```python
from scipy import stats

# Assuming you have a single DataFrame named 'customer_data' with 'Age' and 'Gender' columns

# Separate age data into two groups based on gender
age_men = sd[sd['Gender'] == 'Men']['Age']
age_women = sd[sd['Gender'] == 'Women']['Age']

# Perform a two-sample t-test to compare the average age between men and women
t_statistic, p_value = stats.ttest_ind(age_men, age_women, equal_var=False)

# Define the significance level (alpha)
alpha = 0.05

# Check the p-value against the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
    print("The average age differs between men and women.")
else:
    print("Fail to reject the null hypothesis")
    print("The average age is the same for men and women.")
```

## **Hypothesis Testing 3**

```python
from scipy import stats

# Assuming you have a single DataFrame named 'sales_data'

# Separate sales data into different groups based on product categories
sales_kurta = sd[sd['Category'] == 'Kurta']['Amount']
sales_set = sd[sd['Category'] == 'Set']['Amount']

# Perform an independent t-test to compare sales between product categories
t_statistic, p_value = stats.ttest_ind(sales_kurta, sales_set, equal_var=False)

# Define the significance level (alpha)
alpha = 0.05

# Check the p-value against the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
    print("There is a significant difference in sales between product categories 'Kurta' and 'Set'.")
else:
    print("Fail to reject the null hypothesis")
    print("There is no significant difference in sales between the product categories.")
```

## **Hypothesis Testing 4**

```python
from scipy import stats

# Assuming you have a single DataFrame named 'sales_data'

# Create age groups, for example, 'Young', 'Middle-aged', and 'Elderly'
# You can define your own age groupings based on your dataset
young_sales = sd[sd['Age'] < 30]['Amount']
middle_aged_sales = sd[(sd['Age'] >= 30) & (sd['Age'] < 50)]['Amount']
elderly_sales = sd[sd['Age'] >= 50]['Amount']

# Perform one-way ANOVA to compare sales among different age groups
f_statistic, p_value = stats.f_oneway(young_sales, middle_aged_sales, elderly_sales)

# Define the significance level (alpha)
alpha = 0.05

# Check the p-value against the significance level
if p_value < alpha:
    print("Reject the null hypothesis")
    print("There is a significant difference in the average sales amount among different age groups.")
else:
    print("Fail to reject the null hypothesis")
    print("The average sales amount is the same for different age groups.")
```
